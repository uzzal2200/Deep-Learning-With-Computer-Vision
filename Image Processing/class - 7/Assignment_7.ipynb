## Tast 1:

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
# Read the imge
image=cv.imread('rifat.jpg')
plt.imshow(image)
```

```python

# convert the color space
img_rgb=cv.cvtColor(image,cv.COLOR_BGR2RGB)
image_hsv=cv.cvtColor(img_rgb, cv.COLOR_BGR2HSV)
lower_green = np.array([35, 100, 100])  # Lower bound for green  
upper_green = np.array([85, 255,255])  # Upper bound for green 
thresholded=cv.inRange(image_hsv,lower_green,upper_green)
```

```python
# now plotiing the image binary mask and Original Image
plt.subplot(1,2,1)
plt.imshow(image)
plt.subplot(1,2,2)
plt.imshow(thresholded,cmap='grey')
```

## Task 2

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 

# Load the image
image = cv2.imread('rifat.jpg')  
im_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  
height, width = image.shape[:2]


# Create an empty mask
mask = np.zeros((height, width), dtype=np.uint8)

# Fill the ROI in the mask with white (255)
mask[100:200,200:400] = 255

# Apply the mask on the image
masked_image = cv2.bitwise_and(image, image, mask=mask)

plt.subplot(1,3,1)


plt.imshow(image)
plt.subplot(1,3,2)


plt.imshow(mask, cmap="gray")

plt.subplot(1,3,3)
plt.imshow(masked_image)



```
