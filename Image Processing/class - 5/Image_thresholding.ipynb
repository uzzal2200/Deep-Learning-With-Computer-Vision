```python
import cv2 as cv

import numpy as np 
from matplotlib import pyplot as plt
```

```python
img = cv.imread('D:\Reserach\Deep Learning with Computer Vision\Class - 5\download (1).jpg',cv.IMREAD_GRAYSCALE)
plt.imshow(img)
```

```python
print(img.shape)
```

```python
plt.imshow(img,cmap='gray')
```

       ,   (255)

Binary >threshold (p_val=max_val); <=threshold (p_val=0)

```python
ret , thresholded = cv.threshold(img, 230, 255, cv.THRESH_BINARY)
print(ret)
plt.imshow(thresholded, cmap='gray')
```

Binary INV >threshold (p_val=0); <=threshold (p_val=max_val)

```python
ret, thresholded = cv.threshold(img, 230, 255, cv.THRESH_BINARY_INV)
plt.imshow(thresholded, cmap='gray')

print(ret)
```

Trunc >threshold (p_val=threshold); <=threshold (p_val unchanged)

```python
ret, thresholded = cv.threshold(img, 220, 255, cv.THRESH_TRUNC)

plt.imshow(thresholded, cmap='gray')
print(ret)
```

ToZero >threshold (p_val unchanged); <=threshold (p_val=0)

```python
ret, thresholded = cv.threshold(img, 127, 255,cv.THRESH_TOZERO)
plt.imshow(thresholded, cmap='gray')
print(ret)
```

ToZero INV >threshold (p_val=0/black); <=threshold (p_val unchanged)

```python
ret, thresholded = cv.threshold(img, 127, 255,cv.THRESH_TOZERO_INV)
plt.imshow(thresholded, cmap='gray')
print(ret)
```

```python
ret , thresholded1 = cv.threshold(img, 127, 255, cv.THRESH_BINARY)

ret, thresholded2 = cv.threshold(img, 127, 255, cv.THRESH_BINARY_INV)

ret, thresholded3 = cv.threshold(img, 220, 255, cv.THRESH_TRUNC)

ret, thresholded4 = cv.threshold(img, 127, 255,cv.THRESH_TOZERO)

ret, thresholded5 = cv.threshold(img, 127, 255,cv.THRESH_TOZERO_INV)

titles = ['Orginal Image', 'BINARY','BINARY_INV', 'TRUNC','TOZERO','TOZERO_INV']

images = [img,thresholded1 ,thresholded2,thresholded3,thresholded4,thresholded5]

for i in range(6):
    plt.subplot(2, 3, i+1)  # Arrange the subplots in a 2x3 grid
    plt.imshow(images[i], cmap='gray', vmin=0, vmax=255)  # Display the image
    plt.title(titles[i])  # Add a title for each subplot
    plt.xticks([]), plt.yticks([])  # Remove x and y ticks for cleaner display


plt.show()
```
