```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img=cv.imread("images.jpg",cv.IMREAD_GRAYSCALE)
plt.imshow(img,cmap='gray')
```

```python
res,thresold=cv.threshold(img,127,255,cv.THRESH_BINARY)
res,thresold1=cv.threshold(img,127,255,cv.THRESH_BINARY_INV)
res,thresold2=cv.threshold(img,127,255,cv.THRESH_TRUNC)
res,thresold3=cv.threshold(img,127,255,cv.THRESH_TOZERO)
res,thresold4=cv.threshold(img,127,255,cv.THRESH_BINARY_INV)
```

```python
plt.subplot(1,5,1)
plt.imshow(thresold,cmap='gray')
plt.subplot(1,5,2)
plt.imshow(thresold1,cmap='gray')
plt.subplot(1,5,3)
plt.imshow(thresold2,cmap='gray')
plt.subplot(1,5,4)
plt.imshow(thresold3,cmap='gray')
plt.subplot(1,5,5)
plt.imshow(thresold4,cmap='gray')
```
