```python
import numpy as np 
import cv2 as cv 
import matplotlib.pyplot as plt 
```

```python
img = cv.imread(r'D:\Reserach\Deep Learning with Computer Vision\class - 10\download.png')

```

```python
sobel = cv.Sobel(img, -1,1,0,3)
plt.imshow(sobel,cmap='gray')
plt.title('Sobel')
plt.axis('off')
```

```python
sobel_x = cv.Sobel(img, -1,1,0,3)
sobel_y = cv.Sobel(img, -1,0,1,3)

plt.figure(figsize=(15,15))


plt.subplot(131)
plt.imshow(img,cmap='gray')
plt.title('Orginal')
plt.axis('off')



plt.subplot(132)

plt.imshow(sobel_x, cmap='gray')
plt.title('Vertical sobel filter')
plt.axis('off')

plt.subplot(133)
plt.imshow(sobel_y, cmap='gray')
plt.title('Horizontal sobel filter')
plt.axis('off')
plt.show()
```

```python
laplacian = cv.Laplacian(img,-1, ksize=3)
plt.imshow(laplacian, cmap='gray')
plt.title('Laplacian')
plt.axis('off')

```

```python
# sobel vs laplacain

sobel_x = cv.Sobel(img, -1,1,0,3)
sobel_y = cv.Sobel(img, -1,0,1,3)

laplacian = cv.Laplacian(img, -1)

plt.figure(figsize=(10,8))


plt.subplot(221)
plt.imshow(img, cmap='gray')
plt.title('Orignal')
plt.axis('off')

plt.subplot(222)
plt.imshow(sobel_x, cmap='gray')
plt.title('Horizontal sobel filter')
plt.axis('off')

plt.subplot(223)
plt.imshow(sobel_y, cmap='gray')
plt.title('Horizontal sobel filter')
plt.axis('off')


plt.subplot(224)
plt.imshow(laplacian, cmap='gray')
plt.title('laplacian filter')
plt.axis('off')
plt.show()
```
