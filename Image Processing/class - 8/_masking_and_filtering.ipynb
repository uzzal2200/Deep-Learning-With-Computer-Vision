```python
import cv2 as cv
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
input_image_path = r'D:\Reserach\Deep Learning with Computer Vision\class - 8\images.jpg'

image = cv.imread(input_image_path)
img_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)
img_hsv = cv.cvtColor(image, cv.COLOR_BGR2HSV)

plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title('Orgnal Image')

plt.subplot(1,2,2)
plt.imshow(img_hsv)
plt.title('HSV Formate Image')

```

```python
lower_blue = np.array([100,100,100])
upper_blue = np.array([140,255,255])

thresholded = cv.inRange(img_hsv, lower_blue, upper_blue)

# Display the images using Matplotlib
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title('Orginal Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(thresholded, cmap='gray')
plt.title('Mask')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
img = cv.imread(r'D:\Reserach\Deep Learning with Computer Vision\class - 8\download.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

kernel1 = np.ones((3,3),np.float32)/9
blur_img1 = cv.filter2D(img, -1, kernel1)

kernel2 = np.ones((7,7),np.float32)/49
blur_img2 = cv.filter2D(img, -1, kernel2)

plt.figure(figsize=(15,5))

plt.subplot(131)
plt.imshow(img)
plt.title('Orginal Image')
plt.axis('off')

plt.subplot(132)
plt.imshow(blur_img1)
plt.title('Averaging 3X3')
plt.axis('off')

plt.subplot(133)
plt.imshow(blur_img2)
plt.title('Averaging 7X7')
plt.axis('off')
```
