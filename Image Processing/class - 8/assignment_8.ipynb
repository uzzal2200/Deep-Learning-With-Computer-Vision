Q1. Contribution of filtering in the given applications

Answer
Noise Reduction: Removes unwanted noise using filters like Gaussian, Median, or Low-pass.
Feature Extraction: Highlights critical patterns with Gabor, Wavelet, or LoG filters.
Edge Detection: Identifies object boundaries using Sobel, Canny, or Prewitt filters.
Object Recognition: Enhances object detection with Convolutional, Gabor, or Haar filters.
Enhancing Image: Improves sharpness and contrast using Unsharp Masking, Bilateral, or Histogram Equalization

```python
img = cv.imread(r'D:\Reserach\Deep Learning with Computer Vision\class - 8\image.png')
img = cv.cvtColor(img, cv.COLOR_BGR2RGBA)

kernel1 = np.ones((3,3), np.float32) /9
blur1 = cv.filter2D(img, -1, kernel1)

kernel2 = np.ones((7,7), np.float32)/49
blur2 = cv.filter2D(img, -1, kernel2)

plt.figure(figsize=(15,5))

plt.subplot(131)
plt.imshow(img)
plt.title('Orginal Image')
plt.axis('off')

plt.subplot(132)
plt.imshow(blur1)
plt.title('Averaging 3X3')
plt.axis('off')

plt.subplot(133)
plt.imshow(blur2)
plt.title('Averaging 7X7')
plt.axis('off')



```

```python
blur_image = cv.blur(img, (3,3))
blur_image2 = cv.blur(img, (7,7))

plt.figure(figsize=(15,5))

plt.subplot(131)
plt.imshow(img)
plt.title('Orginal Image ')
plt.axis('off')


plt.subplot(132)
plt.imshow(blur_image)
plt.title('Averaging 3X3')
plt.axis('off')


plt.subplot(133)
plt.imshow(blur_image2)
plt.title('Averaging 7X7')
plt.axis('off')
```
