```python
def white_balance(imag, sf=128):
    # convert the image to float32
    image_float = image.astype(np.float32)
    
    # calculate the average color of the image
    avg_color = np.mean(image_float, axis=(0,1))
    print('average color', avg_color)
    
    # calculate the scaling factors 
    scale = sf / avg_color
    print('scaing factor', scale)
    
    
    # apply the scaling factors to each channel 
    img_balanced = image_float * scale
    print('max and min value', np.max(img_balanced), np.min(img_balanced))
     # Clip the values to be in the valid range [0, 255]
     
    img_balanced = np.clip(img_balanced, 0, 255).astype(np.uint8)

    return img_balanced     
```

```python
image_path = r'D:\Reserach\Deep Learning with Computer Vision\class - 11\image.png'
image = cv.imread(image_path)

balanced_image_1 = white_balance(image)

balanced_image_2 = white_balance(image, sf = 50)


plt.figure(figsize=(10,5))

plt.subplot(131)
plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Original Image ')

plt.subplot(132)
plt.imshow(cv.cvtColor(balanced_image_1, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.title('White Balance Image 1  ')

plt.subplot(133)
plt.imshow(cv.cvtColor(balanced_image_2, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.title('White Balance Image 2 ')

```
