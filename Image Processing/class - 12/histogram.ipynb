```python
import cv2 as cv 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
img = cv.imread(r'image.png')
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
# plt the histogram of the full image
plt.figure(figsize=(8,6))

# flatten the image, plot histogram with black bars 

plt.hist(img.ravel(), bins=256, range=(0,256), color='k')
plt.title('Histogram of Full Image ')
plt.xlabel('Pixel Value ')
plt.ylabel('Frequency')
plt.show()


# plot the histogram of the individual color channels 

plt.figure(figsize=(12,4))
plt.subplot(131)
plt.hist(img_rgb [:, :, 0].ravel(), bins=256, range=(0,256),color='r')
plt.title('Red Channel Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')


plt.subplot(132)
plt.hist(img_rgb[:, :, 1].ravel(), bins=256, range=(0,256), color='g')
plt.title('Green Channel Histogram')
plt.xlabel('Pixel Value ')
plt.ylabel('Frequency')


plt.subplot(133)
plt.hist(img_rgb[:, :, 1].ravel(), bins=256, range=(0,256), color='b')
plt.title('Blue Channel Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

Histogram Equalization

```python
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# Apply histogram equalization 
equalized_image = cv.equalizeHist(img_gray)

plt.figure(figsize=(12,6))

plt.subplot(121)
plt.title('Orginal Image ')
plt.imshow(img_gray, cmap='gray')
plt.axis('off')


plt.subplot(122)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image ')
plt.axis('off')
```

```python
# Histogram compare of Orginal and histogram equalized image 
plt.figure(figsize=(12,4))
plt.subplot(121)
plt.hist(img_gray.ravel(), bins=256, range=(0,256),color='k')
plt.title('Orginal Image ')
plt.xlabel('Pixel Value ')
plt.ylabel('Frequency')

plt.subplot(122)
plt.hist(equalized_image.ravel(), bins=256, range=(0,256), color='k')
plt.title('Histogram Equalized Image ')
plt.xlabel('Pixel Value ')
plt.ylabel('Frequency')
```

```python
ahe = cv.createCLAHE(clipLimit=6.0, tileGridSize=(8,8))
ahe_image = ahe.apply(img_gray)
plt.imshow(ahe_image, cmap='gray')
```

Contrast Limited Adaptive Histogram Equalization - CLAHE

```python
ghe_image = cv.equalizeHist(img_gray)

# Apply CLAHE 
clahe1 = cv.createCLAHE(clipLimit=1.0, tileGridSize=(8,8))
clahe1_img = clahe1.apply(img_gray)

clahe2 = cv.createCLAHE(clipLimit=7.0, tileGridSize=(8,8))
clahe2_img = clahe2.apply(img_gray)

# Plotting the images 
plt.figure(figsize=(10,10))
plt.subplot(221)
plt.imshow(img_gray, cmap='gray')
plt.title('Orginal Image ')
plt.axis('off')

plt.subplot(222)
plt.imshow(ghe_image, cmap='gray')
plt.title('Orginal Histogram Equalization (GHE)')
plt.axis('off')


plt.subplot(223)
plt.imshow(clahe1_img, cmap='gray')
plt.title('CLAHE 1')
plt.axis('off')

plt.subplot(224)
plt.imshow(clahe2_img, cmap='gray')
plt.title('CLAHE 2 ')
plt.axis('off')

plt.tight_layout()
plt.show()
```
