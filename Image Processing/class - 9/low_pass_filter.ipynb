```python
import numpy as np 
import cv2 as cv 
import matplotlib.pyplot as plt 
```

```python
img = cv.imread(r'D:\Reserach\Deep Learning with Computer Vision\class - 9\berry-1.jpg')

img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

plt.imshow(img, cmap='gray')
```

```python
median_blur = cv.medianBlur(img, 5)
plt.imshow(median_blur, cmap='gray')
```

```python
gaussian_blur = cv.GaussianBlur(img, (5,5), 0)

plt.imshow(gaussian_blur, cmap='gray')
```

```python
bilateral_blur = cv.bilateralFilter(img, 5, 75, 75)
plt.imshow(bilateral_blur, cmap='gray')
```

```python
img = cv.imread(r'D:\Reserach\Deep Learning with Computer Vision\class - 9\berry-1.jpg')

img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

avg_blur = cv.blur(img, (5,5))
median_blur = cv.medianBlur(img, 5)

gaussian_blur = cv.GaussianBlur(img, (5,5), 0)

bilateral_blur = cv.bilateralFilter(img, 5, 75, 75)

plt.figure(figsize=(15,5))

plt.subplot(141)
plt.imshow(avg_blur, cmap='gray')
plt.title('Averaaging')
plt.axis('off')

plt.subplot(142)
plt.imshow(median_blur, cmap='gray')
plt.title('Median')
plt.axis('off')

plt.subplot(143)
plt.imshow(gaussian_blur, cmap='gray')
plt.title('Gaussian')
plt.axis('off')

plt.subplot(144)
plt.imshow(bilateral_blur, cmap='gray')
plt.title('Bilateral')
plt.axis('off')


```
