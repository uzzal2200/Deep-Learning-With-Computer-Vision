## 1. Plot a 4*4 grid with original and three flipped images

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread("images.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) 
```

```python
# plot 4*4 grid with original and flipped imagaes
flip_horizontal = cv2.flip(image, 1)  
flip_vertical = cv2.flip(image, 0)  
flip_both = cv2.flip(image, -1)  
```

```python
# ploting
plt.figure(figsize=(12, 6))
plt.subplot(1, 4, 1)
plt.imshow(image)
plt.title("Original")
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(flip_horizontal)
plt.title("Horizontally Flipped")
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(flip_vertical)
plt.title("Vertically Flipped")
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(flip_both)
plt.title("Both Flipped")
plt.axis('off')

plt.show()
```

### 2. Plot 2x3 Grid (Rotated Images)

```python
height, width = image.shape[:2]
center = (width // 2, height // 2)

# Standard OpenCV Rotation
rotate_90_clockwise = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
rotate_90_counterclockwise = cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE)
rotate_180 = cv2.rotate(image, cv2.ROTATE_180)

# Custom Affine Rotation
matrix_45 = cv2.getRotationMatrix2D(center, 45, 1.0)
matrix_135 = cv2.getRotationMatrix2D(center, 135, 1.0)
rotate_45 = cv2.warpAffine(image, matrix_45, (width, height))
rotate_135 = cv2.warpAffine(image, matrix_135, (width, height))
matrix_160= cv2.getRotationMatrix2D(center, 160,1.0)
rotate_160 = cv2.warpAffine(image, matrix_160, (width, height))
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.imshow(rotate_90_clockwise)
plt.title("90° Clockwise")


plt.subplot(2, 3, 2)
plt.imshow(rotate_90_counterclockwise)
plt.title("90° Counterclockwise")


plt.subplot(2, 3, 3)
plt.imshow(rotate_180)
plt.title("180° Rotation")


plt.subplot(2, 3, 4)
plt.imshow(rotate_45)
plt.title("45° Rotation")


plt.subplot(2, 3, 5)
plt.imshow(rotate_135)
plt.title("135° Rotation")

plt.subplot(2, 3, 6)
plt.imshow(rotate_160)
plt.title("160° Rotation")



plt.subplot(2, 3, 6)


plt.show()
```

### 3. Plot 1x3 Grid (Original, Rescaled & Resized)

```python

rescaled_image = cv2.resize(image, None, fx=1, fy=1, interpolation=cv2.INTER_LINEAR)
resized_image = cv2.resize(image, (200, 200), interpolation=cv2.INTER_LINEAR)  # Fixed size (200x200)

plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(image)
plt.title("Original")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(rescaled_image)
plt.title("Rescaled 100%")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(resized_image)
plt.title("Resized (200x200)")
plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image_path = 'images.jpg'  # Replace with your actual image path
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for correct display in Matplotlib

# Flip images
flip_horizontal = cv2.flip(image, 1)  # Horizontal flip
flip_vertical = cv2.flip(image, 0)  # Vertical flip
flip_both = cv2.flip(image, -1)  # Both horizontal & vertical flip

# Plot 4x4 grid with original and flipped images
plt.figure(figsize=(10, 10))

plt.subplot(4, 4, 1)
plt.imshow(image)
plt.title("Original")
plt.axis('off')

plt.subplot(4, 4, 2)
plt.imshow(flip_horizontal)
plt.title("Horizontally Flipped")
plt.axis('off')

plt.subplot(4, 4, 3)
plt.imshow(flip_vertical)
plt.title("Vertically Flipped")
plt.axis('off')

plt.subplot(4, 4, 4)
plt.imshow(flip_both)
plt.title("Both Flipped")
plt.axis('off')

plt.show()
```
