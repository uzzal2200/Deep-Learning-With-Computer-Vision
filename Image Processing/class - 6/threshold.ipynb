```python
import cv2 as cv 
import pandas as pd 
from matplotlib import pyplot as plt 
```

```python
# load the image 
img_path = r'D:\Reserach\Deep Learning with Computer Vision\class - 6\berry-1.jpg'
img = cv.imread(img_path, cv.IMREAD_GRAYSCALE)
plt.imshow(img)
```

```python
img_bgr = cv.imread(img_path)
img_gray = cv.cvtColor(img_bgr, cv.COLOR_BGR2GRAY)

plt.imshow(img_gray)
```

```python
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_rgb = cv.cvtColor(img, cv.COLOR_GRAY2RGB)
plt.imshow(img_rgb)
```

Adaptive Thresholding

```python
img.shape
```

```python
adaptive_mean = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 151,2)

plt.imshow(adaptive_mean, 'gray')
```

```python
adaptive_gaussian = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 2)
plt.imshow(adaptive_gaussian, 'gray')
```

```python
ret1, otsu1 = cv.threshold(img, 160, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)
ret2, otsu2 = cv.threshold(img, 60, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)

plt.subplot(1, 2, 1)
plt.imshow(otsu1, cmap='gray')
# plt.title('Otsu Thresholding (v=160)')

plt.subplot(1, 2, 2)
plt.imshow(otsu2, cmap='gray')
# plt.title('Otsu Thresholding (v=60)')

plt.show()
```
