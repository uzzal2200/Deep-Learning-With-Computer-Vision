### Ans 1
In Adapring Thresolding,`BLOCK size` determines the size of the local region used to compute the threshold for each pixel.
The reson `block size`
1. Center Pixel Calculate: The Threshold for each pixel is compute using its surrounding neighbors within the `block size`.if `block size` is even, there is no exact center pixel, making calculations ambiguous.

2. Symmetric Requirement: An odd `block size` ensures that an equal number of pixels exist in either side of the center pixel, making computations symmetric.
3. kernel Operation in image Processing: Many filtering operations require a symmetric kernal with a well-defined center, which is only possible with an odd `block size`.

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img=cv.imread("berry-1.jpg",cv.IMREAD_GRAYSCALE)
plt.imshow(img,'gray')
```

```python
adapt=cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_MEAN_C,cv.THRESH_BINARY,3,1)
plt.imshow(adapt,'gray')
```

```python
ad=cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,cv.THRESH_BINARY,5,0)
plt.imshow(ad,'gray')
```

```python
titles=['Original','Adapt_Mean','Adapt_Gaussian']
images=[img,adapt,ad]
for i in range(3):
    plt.subplot(1,3,i+1)
    plt.imshow(images[i],'gray')
    plt.title(titles[i])
plt.show()
```

```python
# Otsu's Binarization as global Thresold ,every Global Thresold have the return 2 value one theshold value and threshold image
ret,otsu=cv.threshold(img,127,255,cv.THRESH_BINARY+cv.THRESH_OTSU)
ret,otsu1=cv.threshold(img,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)

plt.subplot(1,2,1)
plt.imshow(otsu,'gray')
plt.subplot(1,2,2)
plt.imshow(otsu1)
```
